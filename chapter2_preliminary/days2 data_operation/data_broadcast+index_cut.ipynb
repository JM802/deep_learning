```python
X=torch.arange(12,dtype=torch.float32).reshape(3,4)
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```

```python
#index==-1:the last element,1:3--1<=index<3(left close,right open)
X[-1], X[1:3]
```

```python
X[1, 2] = 9
X
```

```python
#0:2--same as the front,:--all this dimension
X[0:2, :] = 12
X
```

```python
#deep copy
before = id(Y)
Y = Y + X
id(Y) == before
```

```python
Z = torch.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

```python
#shallow copy(conserve memory space)
before = id(X)
X += Y
id(X) == before
```

```python
A=X.numpy()
B=torch.tensor(A)
type(A),type(B)
```

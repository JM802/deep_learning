```python
import os
os.makedirs(os.path.join(".","data"),exist_ok=True)
data_file=os.path.join(".","data","house.csv")
with open(data_file,"w") as f:
    f.write('NumRooms,Alley,Name,Price\n')  # colunm name
    f.write('NA,Pave,NA,127500\n')  # each line means one example
    f.write('2,NA,Jim,106000\n')
    f.write('4,NA,Anna,178100\n')
    f.write('NA,NA,NA,140000\n')
```

```python
import pandas as pd
data=pd.read_csv(data_file)
```

```python
#calculate each column NaN numbers
na_count=data.isnull().sum()
#get max NaN column name
max_na_col=na_count.idxmax()
#delete this column,axis=1:delete column,default:inplace=false
data_drop=data.drop(max_na_col,axis=1)
```

```python
print(f"缺失值最多的列是：{max_na_col}（缺失数：{na_count[max_na_col]}）")
print(data_drop)
```

```python
data_drop=pd.get_dummies(data_drop, dummy_na=True)
data_drop
```

```python
import torch
x=torch.tensor(data_drop.to_numpy(dtype=float))
x
```

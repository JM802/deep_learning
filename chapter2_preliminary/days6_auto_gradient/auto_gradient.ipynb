```python
import torch

x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)  # when we open requires_grad,we can track the process of the tensor calculate
x.grad  # 默认值是None
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
#the grad is a attribute of tensor
x.grad
```

```python
x.grad == 4 * x
```

```python
# 在默认情况下，PyTorch会累积梯度，我们需要清除之前的值
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y=x*x#y is a tensor
#pytorch default:Derivative of a Scalar with Respect to a Tensor(标量对张量求导)

#equivalent to y.backward(torch.ones(len(x)))
y.sum().backward()#equivalent to scalar, converting a tensor to a scalar lets us compute gradients effortlessly
x.grad
```

```python
x.grad.zero_()
y=x*x
u=y.detach()#untrack the y=x*x,u can be seen as a constant
z=u*x

z.sum().backward()
x.grad==u
```

```python
x.grad.zero_()
y.sum().backward()#y is not like u,y is tracked
x.grad == 2 * x
```

```python
#wwe can figure out c=k*a
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward() 
a.grad == d / a
```

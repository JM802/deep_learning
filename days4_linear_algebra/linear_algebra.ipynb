```python
A = torch.arange(20,dtype=torch.float32).reshape(5, 4)
A
```

```python
A.T
```

```python
sum_A = A.sum(axis=1, keepdims=True)
sum_A
```

```python
#broadcast,sum_A copy the column to 5*4
A / sum_A
```

```python
#cumulative in target axis
A.cumsum(axis=0)
```

```python
#L2 norm:ordinary distance
u = torch.tensor([3.0, -4.0])
torch.norm(u)
```

```python
# L1 norm: the sum of absolute values (also known as Manhattan distance for vectors)
torch.abs(u).sum()
```
